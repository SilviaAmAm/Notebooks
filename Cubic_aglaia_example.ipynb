```python
from aglaia import aglaia
import tensorflow as tf
import numpy as np
```

```python
x = np.linspace(-2.0, 2.0, 8)
X = np.reshape(x, (len(x), 1))
y = x**3
```

```python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

```python
estimator = aglaia.MRMP(hidden_layer_sizes=(5, 5, 5), learning_rate=0.01, iterations=35000, l2_reg = 0, tf_dtype = 32, scoring_function="rmse")
```

```python
estimator.fit(X, y)
```

```python
x_test = np.linspace(-1.5, 1.5, 15)
X_test = np.reshape(x_test, (len(x_test), 1))
y_test = x_test**3
y_pred = estimator.predict(X_test)
```

```python
plt.scatter(x_test,y_pred)
plt.scatter(x_test,y_test)
plt.show()
```

```python
estimator.plot_cost()
```

```python
err = y_test - np.reshape(y_pred, (y_pred.shape[0],))
print(err)
```

```python
y_pred_row = np.reshape(y_pred, (y_pred.shape[0],))
np.testing.assert_array_almost_equal(y_test, y_pred_row, decimal=1)
```
