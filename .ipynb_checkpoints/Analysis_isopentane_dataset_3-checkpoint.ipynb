This notebook does the same analysis that is in Analysis_isopentane_dataset3 but with the dftb energies.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = np.load("/Users/walfits/Desktop/CN_isopentane_forces.npz")
xyz = data["arr_0"]
zs = data["arr_1"][0]
ene = data["arr_2"]/4.184 - (-2800)
print(xyz.shape, zs.shape, ene.shape)
```

```python
# Looking at which H are primary, secondary and tertiary and their index in VMD
h_idx = [2, 3, 4, 5, 7, 9, 11, 12, 13, 14, 15, 16]
h = [1, 1, 1, 2, 2, 3, 1, 1, 1, 1, 1, 1]
```

```python
n_samples = xyz.shape[0]
n_atoms = xyz.shape[1]
```

```python
# Calculating the distance between the C of the CN and all the hydrogens
ch_distance = np.zeros((n_samples, 12))
ch_1 = []
ch_2 = []
ch_3 = []

dist_1 = []
dist_2 = []
dist_3 = []

for i in range(n_samples):
    h_counter = 0
    for j in range(n_atoms):
        if zs[j] == 1:
            dist_vec = xyz[i][-2] - xyz[i][j]
            dist = np.linalg.norm(dist_vec)
            ch_distance[i][h_counter] = dist
            
            # Separating samples where the C-H distance is below 3 A for each type of H
            if dist <= 4.0:
                if h[h_counter] == 1:
                    ch_1.append(i)
                    dist_1.append(dist)
                    
                elif h[h_counter] == 2:
                    ch_2.append(i)
                    dist_2.append(dist)
                    
                elif h[h_counter] == 3:
                    ch_3.append(i)
                    dist_3.append(dist)
                    
                    
            h_counter += 1
```

```python
ene_1 = []
ene_2 = []
ene_3 = []

for idx in ch_1:
    ene_1.append(ene[idx])
for idx in ch_2:
    ene_2.append(ene[idx])
for idx in ch_3:
    ene_3.append(ene[idx])
```

```python
print(len(ene_1), len(dist_1))
```

```python
plt.scatter(dist_1, ene_1, marker='.', alpha=0.4)
plt.xlabel("Primary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.8, 2.5)
# plt.ylim(-2600, -2540)
# plt.savefig("energies_1_dftb.png", dpi=200)
plt.show()
```

```python
plt.scatter(dist_2, ene_2, marker='.', alpha=0.4)
plt.xlabel("Primary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.8, 2.5)
# plt.ylim(-420, -300)
# plt.savefig("energies_1_dftb.png", dpi=200)
plt.show()
```

```python
plt.scatter(dist_3, ene_3, marker='.', alpha=0.4)
plt.xlabel("Primary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.8, 2.5)
# plt.ylim(-420, -300)
# plt.savefig("energies_1_dftb.png", dpi=200)
plt.show()
```
