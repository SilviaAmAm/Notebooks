In this notebook I plot the energy of all the structures that have a C-H distance below a threshold of 2 Angstroms, to see if there is still a stabilised transitions state.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
r = []  # List of files to be joined together
subdirs = [x[0] for x in os.walk("/Volumes/Transcend/data_sets/CN_isobutane_model/pruned_cn_isobutane/training")]
for subdir in subdirs:
    files = next(os.walk(subdir))[2]

    for file in files:
        isTrajectory = file.find(".xyz")
        if isTrajectory >= 0:
            r.append(subdir + "/" + file)
```

```python
n_samples = len(r)
n_atoms = 19
print(n_samples)
```

```python
xyz = np.zeros((n_samples, n_atoms, 3))
zs = np.zeros((n_atoms,))
ene = np.zeros((n_samples,))
```

```python
ene = np.loadtxt("/Volumes/Transcend/data_sets/CN_isobutane_model/pruned_cn_isobutane/training/properties_kjmol.txt", usecols=[1])
print(ene.shape)
```

```python
elements = {"C": 6, "N":7, "H":1}
inputFile = open(r[0], 'r')
line = next(inputFile)
line = next(inputFile)
atom_counter = 0
for line in inputFile:
    line = line.strip()
    lineSplit = line.split("\t")
    zs[atom_counter] = elements[lineSplit[0]]
    atom_counter += 1
print(zs)
```

```python
for i in range(n_samples):
    inputFile = open(r[i], 'r')
    line = next(inputFile)
    line = next(inputFile)
    atom_counter = 0
    for line in inputFile:
        line = line.strip()
        lineSplit = line.split("\t")
        for j in range(1, 4):
            xyz[i][atom_counter][j-1] = float(lineSplit[j])
        atom_counter += 1
```

```python
# Looking at which H are primary, secondary and tertiary and their index in VMD
h_idx = [2, 3, 4, 5, 7, 9, 11, 12, 13, 14, 15, 16]
h = [1, 1, 1, 2, 2, 3, 1, 1, 1, 1, 1, 1]
```

```python
# Calculating the distance between the C of the CN and all the hydrogens
ch_distance = np.zeros((n_samples, 12))
ch_1 = []
ch_2 = []
ch_3 = []

dist_1 = []
dist_2 = []
dist_3 = []

mystery_structures_1 =  []
mystery_structures_2 =  []
mystery_structures_3 =  []

for i in range(n_samples):
    h_counter = 0
    for j in range(n_atoms):
        if zs[j] == 1.0:
            dist_vec = xyz[i][-2] - xyz[i][j]
            dist = np.linalg.norm(dist_vec)
            ch_distance[i][h_counter] = dist
            
            # Separating samples where the C-H distance is below 3 A for each type of H
            if dist <= 4.0:
                if h[h_counter] == 1:
                    ch_1.append(i)
                    dist_1.append(dist)
                    if dist > 1.8 and dist < 2.4:
                        if ene[i] < -380.0:
                            mystery_structures_1.append(i)
                elif h[h_counter] == 2:
                    ch_2.append(i)
                    dist_2.append(dist)
                    if dist > 1.8 and dist < 2.3:
                        if ene[i] < -330.0:
                            mystery_structures_2.append(i)
                elif h[h_counter] == 3:
                    ch_3.append(i)
                    dist_3.append(dist)
                    if dist > 1.9 and dist < 2.3:
                        if ene[i] < -260.0:
                            mystery_structures_3.append(i)
                    
            h_counter += 1
```

```python
ene_1 = []
ene_2 = []
ene_3 = []

for idx in ch_1:
    ene_1.append(ene[idx])
for idx in ch_2:
    ene_2.append(ene[idx])
for idx in ch_3:
    ene_3.append(ene[idx])
```

```python
print(len(ene_1), len(dist_1))
```

```python
plt.scatter(dist_1, ene_1, marker='.', alpha=0.6)
plt.xlabel("Primary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.8, 2.5)
# plt.ylim(-420, -300)
plt.savefig("/Users/walfits/Desktop/analysis_data/energies_1.png", dpi=200)
plt.show()
```

```python
plt.scatter(dist_2, ene_2, marker='.', alpha=0.6)
plt.xlabel("Secondary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.7, 2.5)
# plt.ylim(-420, -300)
plt.savefig("/Users/walfits/Desktop/analysis_data/energies_2.png", dpi=200)
plt.show()
```

```python
plt.scatter(dist_3, ene_3, marker='.', alpha=0.6)
plt.xlabel("Tertiary C-H distance (Angstroms)")
plt.ylabel("Energy (kJ/mol)")
# plt.xlim(1.8, 2.5)
# plt.ylim(-420, -250)
plt.savefig("/Users/walfits/Desktop/analysis_data/energies_3.png", dpi=200)
plt.show()
```

```python
# Writing the mystery strucures to a file for VMD

f = open("mystery_structures_1.xyz", 'w')

f.write("19\n")
f.write("\n")

for idx in mystery_structures_1:
    mystery_xyz = xyz[idx]
    for i in range(mystery_xyz.shape[0]):
        f.write(str(zs[i]))
        for j in range(3):
            f.write("\t" + str(xyz[idx][i][j]))
        f.write("\n")
        
    f.write("19\n")
    f.write("\n")   
```
