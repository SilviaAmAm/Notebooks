```python
from aglaia import symm_funct
from aglaia.tests import np_symm_funct
from aglaia.tests import tensormol_symm_funct
import numpy as np
import tensorflow as tf
```

```python
input_data = "/Volumes/Transcend/repositories/Aglaia/aglaia/tests/data_test_acsf.npz"
data = np.load(input_data)
```

```python
xyzs = data["arr_0"]
zs = data["arr_1"]
elements = data["arr_2"]
element_pairs = data["arr_3"]
```

```python
acsf_tf_t = symm_funct.generate_parkhill_acsf(xyzs, zs, elements, element_pairs)
acsf_tm_t = tensormol_symm_funct.tensormol_acsf(xyzs, zs, elements, element_pairs)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
acsf_tf = sess.run(acsf_tf_t)
acsf_tm = sess.run(acsf_tm_t)
```

```python
print(acsf_tf.shape)
print(acsf_tm.shape)
```

```python
acsf_np = np_symm_funct.generate_acsf(xyzs, zs, elements, element_pairs)
print(acsf_np.shape)
```

```python
acsf_tf_2 = np.reshape(acsf_tf, acsf_tm.shape)
print(acsf_tf_2.shape)
acsf_np_2 = np.reshape(acsf_np, acsf_tm.shape)
print(acsf_np_2.shape)
```

```python
acsf_tf_sort = np.sort(acsf_tf_2, axis=-1)
acsf_np_sort = np.sort(acsf_np_2, axis=-1)
acsf_tm_sort = np.sort(acsf_tm, axis=-1)
```

```python
import pandas as pd
```

```python
df_tf = pd.DataFrame(acsf_tf_sort)
df_tm = pd.DataFrame(acsf_tm_sort)
df_np = pd.DataFrame(acsf_np_sort)
```

```python
df_tf["norm"] = np.sqrt(np.square(df_tf).sum(axis=1))
```

```python
df_tm["norm"] = np.sqrt(np.square(df_tm).sum(axis=1))
df_np["norm"] = np.sqrt(np.square(df_np).sum(axis=1))
```

```python
df_tf_sort = df_tf.sort_values(by=['norm'])
df_tm_sort = df_tm.sort_values(by=['norm'])
df_np_sort = df_np.sort_values(by=['norm'])
```
