```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
sns.set()
x = np.array([1, 5])
y = 0.2* x**2 + 4.5
line_2 = plt.scatter(x, y, c=sns.xkcd_rgb["denim blue"])
plt.xlabel("Time step")
plt.ylabel("Energy (kJ/mol)")
plt.xlim((0,6))
plt.ylim((3.0, 13.0))
plt.savefig("example1.png", dpi=300)
plt.show()
```

```python
x = np.array([1, 3, 5])
y = 0.2* x**2 + 4.5
line_2 = plt.scatter(x, y, c=sns.xkcd_rgb["denim blue"])
plt.xlabel("Time step")
plt.ylabel("Energy (kJ/mol)")
plt.xlim((0,6))
plt.ylim((3.0, 13.0))
plt.savefig("example2.png", dpi=300)
plt.show()
```

```python
x = np.array([1, 3, 5])
y = 0.2* x**2 + 4.5
x_1 = np.arange(0,6, 0.2)
y_1 = 0.2* x_1**2 + 4.5
line_2 = plt.scatter(x, y, c=sns.xkcd_rgb["denim blue"])
line_2 = plt.plot(x_1, y_1, c=sns.xkcd_rgb["medium green"])
plt.xlabel("Time step")
plt.ylabel("Energy (kJ/mol)")
plt.xlim((0,6))
plt.ylim((3.0, 13.0))
plt.savefig("example3.png", dpi=300)
plt.show()
```
