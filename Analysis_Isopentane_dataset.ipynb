In this notebook I calculate the distance from the cyano Carbon to all of the hydrogens in the system to try and estimate how many of each primary, secondary and tertiary abstractions I have.

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
r = []  # List of files to be joined together
subdirs = [x[0] for x in os.walk("/Volumes/Transcend/data_sets/CN_isobutane_model/geoms_2/training")]
for subdir in subdirs:
    files = next(os.walk(subdir))[2]

    for file in files:
        isTrajectory = file.find(".xyz")
        if isTrajectory >= 0:
            r.append(subdir + "/" + file)
```

```python
n_samples = len(r)
n_atoms = 19
```

```python
xyz = np.zeros((n_samples, n_atoms, 3))
zs = np.zeros((n_atoms,))
```

```python
elements = {"C": 6, "N":7, "H":1}
inputFile = open(r[0], 'r')
line = next(inputFile)
line = next(inputFile)
atom_counter = 0
for line in inputFile:
    line = line.strip()
    lineSplit = line.split("\t")
    zs[atom_counter] = elements[lineSplit[0]]
    atom_counter += 1
print(zs)
```

```python
for i in range(n_samples):
    inputFile = open(r[i], 'r')
    line = next(inputFile)
    line = next(inputFile)
    atom_counter = 0
    for line in inputFile:
        line = line.strip()
        lineSplit = line.split("\t")
        for j in range(1, 4):
            xyz[i][atom_counter][j-1] = float(lineSplit[j])
        atom_counter += 1
```

```python
# Looking at which H are primary, secondary and tertiary and their index in VMD
h_idx = [2, 3, 4, 5, 7, 9, 11, 12, 13, 14, 15, 16]
h = [1, 1, 1, 2, 2, 3, 1, 1, 1, 1, 1, 1]
```

```python
# Calculating the distance between the C of the CN and all the hydrogens
ch_distance = np.zeros((n_samples, 12))
weird_structures_idx = []

for i in range(n_samples):
    h_counter = 0
    for j in range(n_atoms):
        if zs[j] == 1.0:
            dist_vec = xyz[i][-2] - xyz[i][j]
            dist = np.linalg.norm(dist_vec)
            ch_distance[i][h_counter] = dist
            h_counter += 1
            # Test to understand WTF are the structures with 1.4 A CH distance
            if dist > 1.3 and dist <= 1.4:
                weird_structures_idx.append(i)
```

```python
bins = np.arange(0.6, 12.0, 0.1)
primary_hist = []
secondary_hist = []
tertiary_hist = []
```

```python
for i in range(len(h)):
    hist, bin_edges = np.histogram(ch_distance[:, i], bins=bins, density=False)
    if h[i] == 1:
        primary_hist.append(hist)
    elif h[i] == 2:
        secondary_hist.append(hist)
    else:
        tertiary_hist.append(hist)
```

```python
final_primary_hist = primary_hist[0]
for i in range(1, len(primary_hist)):
    final_primary_hist += primary_hist[i]

final_secondary_hist = secondary_hist[0]
for i in range(1, len(secondary_hist)):
    final_secondary_hist += secondary_hist[i]
    
final_tertiary_hist = tertiary_hist[0]
```

```python
plt.plot(bins[1:], final_primary_hist)
plt.xlim(0,3)
plt.ylim(0,1000)
plt.xlabel("Primary C-H distances (Angstroms)")
plt.ylabel("Occurrences")
plt.savefig("/Users/walfits/Desktop/analysis_data/bins_1.png", dpi=200)
plt.show()
```

```python
plt.plot(bins[1:], final_secondary_hist)
plt.xlim(0,3)
plt.ylim(0,1000)
plt.xlabel("Secondary C-H distances (Angstroms)")
plt.ylabel("Occurrences")
plt.savefig("/Users/walfits/Desktop/analysis_data/bins_2.png", dpi=200)
plt.show()
```

```python
plt.plot(bins[1:], final_tertiary_hist)
plt.xlim(0,3)
plt.ylim(0,1000)
plt.xlabel("Tertiary C-H distances (Angstroms)")
plt.ylabel("Occurrences")
plt.savefig("/Users/walfits/Desktop/analysis_data/bins_3.png", dpi=200)
plt.show()
```
